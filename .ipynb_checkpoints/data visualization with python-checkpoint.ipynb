```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[1,4,9]
z=[10,5,0]
f=[7,6,9]
t=[7,8,5]


plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,f)
plt.plot(x,t)

plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y","this is z","this is f","this is t"])
plt.show()
```
